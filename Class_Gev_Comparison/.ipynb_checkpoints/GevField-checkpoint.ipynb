```python
import numpy as np
import matplotlib.pylab as plt
from scipy.interpolate import interp1d
```

# Stress tensor powerspectrum

```python
pi_z_100=np.loadtxt("./lcdm_pk000_pi_k.dat")
pi_z_10=np.loadtxt("./lcdm_pk001_pi_k.dat")
pi_z_3=np.loadtxt("./lcdm_pk002_pi_k.dat")
pi_z_0=np.loadtxt("./lcdm_pk003_pi_k.dat")

piv_z_100=np.loadtxt("./lcdm_pk000_pi_v_k.dat")
piv_z_10=np.loadtxt("./lcdm_pk001_pi_v_k.dat")
piv_z_3=np.loadtxt("./lcdm_pk002_pi_v_k.dat")
piv_z_0=np.loadtxt("./lcdm_pk003_pi_v_k.dat")

phi_z_100=np.loadtxt("./lcdm_pk000_phi.dat")
phi_z_10=np.loadtxt("./lcdm_pk001_phi.dat")
phi_z_3=np.loadtxt("./lcdm_pk002_phi.dat")
phi_z_0=np.loadtxt("./lcdm_pk003_phi.dat")

phib_z_100=np.loadtxt("./lcdm_pk000_phi_n.dat")
phib_z_10=np.loadtxt("./lcdm_pk001_phi_n.dat")
phib_z_3=np.loadtxt("./lcdm_pk002_phi_n.dat")
phib_z_0=np.loadtxt("./lcdm_pk003_phi_n.dat")

hiclass=np.loadtxt("./../kessence_gev_cs0z100.dat")
class_file=np.loadtxt("./../class_tk.dat")
class_phi_z_0=np.loadtxt("./lcdm_class_z1_tk.dat")
class_phi_z_100=np.loadtxt("./lcdm_class_z2_tk.dat")

# pi_z_0=np.loadtxt("./lcdm_pk002_pi_k.dat")
# piv_z_3=np.loadtxt("./lcdm_pk000_pi_k_v.dat")
# piv_z_1=np.loadtxt("./lcdm_pk001_pi_k_v.dat")
# piv_z_0=np.loadtxt("./lcdm_pk002_pi_k_v.dat")
# pi_class=np.loadtxt("./kessence.dat")
```

# \pi_k dynamics 

```python
As=2.19*10**-9;
h=0.67;
kp=0.05/h; 
ns=0.96;
P_class_dimensionless=As*(hiclass[:,1]**2)*((hiclass[:,0]/kp)**(ns-1.));
plt.plot(hiclass[:,0],P_class_dimensionless[:],color="red",label=r"$\pi_k$ hiclass z=100 ")
# plt.plot(Gev1[:,0],Gev1[:,2],color="blue",label="pi prime field transfer fucntion")
plt.plot(pi_z_100[:,0],pi_z_100[:,1],color="black",label=r"$\pi_k$ Gev z=100 ")
plt.plot(pi_z_10[:,0],pi_z_10[:,1],color="blue",label=r"$\pi_k$ Gev z=10 ")
plt.plot(pi_z_3[:,0],pi_z_3[:,1],color="magenta",label=r"$\pi_k$ Gev z=3 ")
plt.plot(pi_z_0[:,0],pi_z_0[:,1],color="purple",label=r"$\pi_k$ Gev z=0 ")
# plt.plot(pi_class[:,0],P_class_dimensionless[:],color="green",label=r" z=3, hiclass")
plt.legend(bbox_to_anchor=(0., 0.25, 0.4, .102), loc=1,ncol=1,fontsize=8, mode="expand", borderaxespad=0.)
plt.xscale('log')
plt.yscale("log")

# plt.yscale('log')
plt.xlabel("k[h/Mpc]")
plt.ylabel(r"$P_{\pi}[Mpc^3]$")
plt.xlim(0.01,5)
plt.ylim(1.e-17,1.e-1)

# plt.savefig('field_derivative_comparison.jpg',bbox_inches='tight')
plt.show()
```

# \pi_v dynamics 

```python
As=2.19*10**-9;
h=0.67;
kp=0.05/h; 
ns=0.96;
P_class_dimensionless=As*(hiclass[:,2]**2)*((hiclass[:,0]/kp)**(ns-1.));
plt.plot(hiclass[:,0],P_class_dimensionless[:],color="red",label=r"$\pi_k$ hiclass z=100 ")
plt.plot(piv_z_100[:,0],piv_z_100[:,1],color="black",label=r"$\pi_v$ Gev z=100 ")
plt.plot(piv_z_10[:,0],piv_z_10[:,1],color="blue",label=r"$\pi_v$ Gev z=10 ")
plt.plot(piv_z_3[:,0],piv_z_3[:,1],color="magenta",label=r"$\pi_v$ Gev z=3 ")
plt.plot(piv_z_0[:,0],piv_z_0[:,1],color="purple",label=r"$\pi_v$ Gev z=0 ")
# plt.plot(pi_class[:,0],P_class_dimensionless[:],color="green",label=r" z=3, hiclass")
plt.legend(bbox_to_anchor=(0., 0.25, 0.4, .102), loc=1,ncol=1,fontsize=8, mode="expand", borderaxespad=0.)
plt.xscale('log')
plt.yscale("log")

# plt.yscale('log')
plt.xlabel("k[h/Mpc]")
plt.ylabel(r"$P_{\pi_v}[Mpc^3]$")
plt.xlim(0.01,5)
plt.ylim(1.e-27,1.e-1)

# plt.savefig('field_derivative_comparison.jpg',bbox_inches='tight')
plt.show()
```
